```python
import os 
import numpy as np
```

```python
test_path = "/home/bahao/mse/code/detect/data/raw_data/test"
model_path =  "runs/detect/train18/weights/best.pt"
```

```python
from ultralytics import YOLO

model = YOLO(model_path)

model.predict("/home/bahao/mse/code/detect/data/raw_data/test/0aea0556667f66d81013262223.jpg")
```

```python
origin_files =  os.listdir(test_path)
```

```python
result_json = []
```

```python
for origin_file in origin_files:
    # print(origin_file)
    detect_result = model.predict(f"{test_path}/{origin_file}",  conf=0.33, iou=0.75, verbose=False)[0]
    xyxy_np = np.array(detect_result.boxes.xyxy.cpu())
    for item in xyxy_np.tolist():
        print(item)
        result_json.append({
            "name": origin_file,
            "bbox": item
        })
    
```

```python
import json


with open("res.json", "w") as resf:
    json.dump(result_json, resf)
```
